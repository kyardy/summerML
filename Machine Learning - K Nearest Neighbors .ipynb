```python
import numpy as np
from sklearn import preprocessing, cross_validation,neighbors,svm
import pandas as pd

df = pd.read_csv('testData.csv')

df.replace ('-', -99999, inplace=True) # replacing anything with a -
#df.drop(['GPS Time'],1,inplace=True)

#feature_col_names=['GPS_Speed','Absolute_Throttle_Position_B','Average_trip_speed(moving only)','Engine_Load(%)','Engine_Load(Absolute)(%)','Engine_RPM','Fuel flow rate/hour(gal/hr)','Fuel flow rate/minute(gal/min)','Relative Throttle Position(%)','Speed(mph)','Throttle Position(Manifold)(%)','Torque(Nm)']
#predicted_class_names = ['Normal Driving]


X = np.array(df.drop(['Normal Driving'],1))   #features 
y = np.array(df['Normal Driving']) #Labels

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y,test_size=0.3) #Shuffling data into testing and training chunks

clf = neighbors.KNeighborsClassifier()
clf.fit(X_train,y_train)

accuracy = clf.score(X_test, y_test)
print(accuracy)


example_measures = np.array([30, 20, 35, 80, 80,3000,3.00,1,30,60,40,200])
example_measures = example_measures.reshape(1,-1)

prediction = clf.predict(example_measures)
print(prediction)

#clf = svm.SVC()
#clf.fit(X_train, y_train)
#accuracy = clf.score(X_test, y_test)
#print(accuracy)



#raw_data = open(filename,'rt')
#data = numpy.loadtxt(raw_data,delimiter=",")
#print(data.shape)
```
